```python
import random
import time

# 随机生成数组
def random_array(n):
    rand_array=random.choices(range(100),k=n)
    return rand_array

array_100 = random_array(100)
array_1000 = random_array(1000)
array_10000 = random_array(10000)

# 选择排序
def Selection_sort(a):
    for i in range (len(a)-1):
        min=i
        for j in range(i+1,len(a)):
            if a[j]<=a[min]:
                min=j           
        a[i],a[min] = a[min],a[i]
    return a

# 归并排序
def merge_sort(arr):

    if len(arr) <= 1:
        return arr
    # 找到中间位置
    mid = len(arr) // 2
    # 分别对左右两半进行归并排序
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    # 合并两个已排好序的数组
    return merge(left, right)

def merge(left, right):
    # 初始化一个空列表存放结果
    result = []
    # 当两个数组都未空时，比较并合并
    # 每次循环会处理left或right中最小的元素，并使对应指针向前移动一位，直至一方为空
    while left and right:
        # 将较小的元素从对应列表中弹出，添加到result结果中
        if left[0] < right[0]:
            result.append(left.pop(0))
        else:
            result.append(right.pop(0))
    # 处理剩余元素，将未空数组的剩余元素全添加到result中
    result.extend(left or right)
    return result

# 使用python内置的sorted()函数
def built_in_sort(a):
    return sorted(a)

# 求出各排序所需时间的函数
def Selection_sort_time(a):
    start_time = time.time()
    Selection_sort(a)
    print("Selection_sort：",time.time() - start_time)

def merge_sort_time(a):
    start_time = time.time()
    merge_sort(a)
    print("merge_sort_time：",time.time() - start_time)

def built_in_sort_time(a):
    start_time = time.time()
    built_in_sort(a)
    print("built_in_sort_time：",time.time() - start_time)

# 打印各数组排序的时间
print("array_100排序比较：")
Selection_sort_time(array_100)
merge_sort_time(array_100)
built_in_sort_time(array_100)
print("\n")

print("array_1000排序比较：")
Selection_sort_time(array_1000)
merge_sort_time(array_1000)
built_in_sort_time(array_1000)
print("\n")

print("array_10000排序比较：")
Selection_sort_time(array_10000)
merge_sort_time(array_10000)
built_in_sort_time(array_10000)
```
